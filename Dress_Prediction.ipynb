```python
from tensorflow.keras.models import load_model
from skimage.io import imread,imshow
from skimage.transform import resize
```

```python
# Load the model
model = load_model('fashion_model_1.h5')
```

```python
img_path="c.jpg"
image=imread(img_path)
image
```

```python
imshow(image)
```

```python
image=resize(image,(224,224))
```

```python
image=image.reshape(1,224,224,3)
```

```python
prediction=model.predict(image)
```

```python
ind=prediction.argmax(axis=1)
ind
```

```python

categories=["Casual","Formal","Sports Wear"]
```

```python
categories.sort()
categories
```

```python
class_label=categories[ind[0]]
class_label
```

```python
categories=["Casual","Formal","Sports Wear"]
img_path="c.jpg"
image=imread(img_path)
image=resize(image,(224,224))
image=image.reshape(1,224,224,3)
prediction=model.predict(image)
ind=prediction.argmax(axis=1)
class_label=categories[ind[0]]
class_label
```
